# Colab Environment Setup
Run this notebook **once per Colab session** before executing any other project notebooks.  
It will:
1. Verify the Python version
2. Clone (or refresh) the team GitHub repository using a Personal Access Token
3. Install all required packages from `requirements.txt`
4. Create the expected local directory structure
5. Verify imports and confirm the environment is ready

>**Note:** Colab resets the runtime environment after each session. You must re-run this notebook at the start of every new session.

## 1) Verify Python Version

This project requires **Python 3.10 or higher**. The cell below confirms the active interpreter.

````python
import sys

print(f"Python version: {sys.version}")
print(f"Python executable: {sys.executable}")

major, minor = sys.version_info.major, sys.version_info.minor
if (major, minor) < (3, 10):
    raise EnvironmentError(
        f"Python 3.10+ is required, but found {major}.{minor}. "
        "Go to Runtime > Change runtime type and select a newer runtime."
    )
else:
    print(f"\n Python {major}.{minor} — version requirement satisfied.")
````

## 2) Configure GitHub Credentials

Enter your **GitHub Personal Access Token (PAT)** and repository details below.  
Your PAT needs at least `repo` (read) scope. Generate one at:  
https://github.com/settings/tokens

>**Security:** Never hard-code your token. Enter it only in the input prompt below — it will not be stored in the notebook or committed to GitHub.

````python
import getpass

# ── Repository configuration ──────────────────────────────────────────────────
GITHUB_USERNAME = "payalpt"   # ← replace with your GitHub username
REPO_NAME       = "AAI-590-capstone-canadian-health"        # ← replace with your exact repo name
BRANCH          = "main"                   # ← change if your default branch differs
# ──────────────────────────────────────────────────────────────────────────────
# https://github.com/payalpt/AAI-590-capstone-canadian-health.git #
# Prompt securely — input is masked and not saved
GITHUB_TOKEN = getpass.getpass("Enter your GitHub Personal Access Token (input hidden): ")

REPO_URL = f"https://{GITHUB_USERNAME}:{GITHUB_TOKEN}@github.com/{GITHUB_USERNAME}/{REPO_NAME}.git"
print(f"\nRepository target: https://github.com/{GITHUB_USERNAME}/{REPO_NAME} (branch: {BRANCH})")
print("Token received — credentials configured.")
````

## 3) Clone or Refresh the Repository

If the repository has already been cloned in this session, it will be pulled to the latest commit instead of re-cloned.

````python
import os
import subprocess

REPO_DIR = f"/content/{REPO_NAME}"

def run(cmd, cwd=None):
    """Run a shell command and print output; raise on non-zero exit."""
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True, cwd=cwd)
    if result.stdout:
        print(result.stdout.strip())
    if result.returncode != 0:
        raise RuntimeError(f"Command failed:\n{result.stderr.strip()}")

if os.path.isdir(REPO_DIR):
    print(f"Repository already exists at {REPO_DIR} — pulling latest changes...")
    # Update the remote URL in case the token changed
    run(f"git remote set-url origin {REPO_URL}", cwd=REPO_DIR)
    run(f"git fetch origin", cwd=REPO_DIR)
    run(f"git checkout {BRANCH}", cwd=REPO_DIR)
    run(f"git pull origin {BRANCH}", cwd=REPO_DIR)
    print(f"\n Repository refreshed to latest commit on '{BRANCH}'.")
else:
    print(f"Cloning repository into {REPO_DIR} ...")
    run(f"git clone --branch {BRANCH} {REPO_URL} {REPO_DIR}")
    print(f"\n Repository cloned successfully.")

# Show the latest commit for confirmation
run("git log --oneline -5", cwd=REPO_DIR)
````

## 4) Persist the Environment on Google Drive

By default, Colab discards all installed packages when the runtime disconnects. Mounting Google Drive and installing packages into a Drive-backed folder means:

> **Drive path used:** `My Drive/colab_env/aai590`  
> You will be prompted to authenticate with your Google account when Drive mounts.

### Using the Persistent Environment in Other Notebooks

Add the following two lines to the **top of every other project notebook** (after any existing `import sys` statements). This is the only setup needed in subsequent notebooks once the packages are installed here:

```python
import sys
sys.path.insert(0, "/content/drive/MyDrive/AAI-590-Capstone-Project/colab_env")
# All packages installed by 00_colab_setup.ipynb are now importable
```

Google Drive must still be mounted in each new session (`drive.mount(...)`) before the path insert will work. The two-line snippet above handles everything else — no reinstall required.

````python
import os
from google.colab import drive

# ── Mount Google Drive ────────────────────────────────────────────────────────
DRIVE_MOUNT   = "/content/drive"
drive.mount(DRIVE_MOUNT, force_remount=False)   # force_remount=False skips re-auth if already mounted

# ── Persistent environment directory on Drive ─────────────────────────────────
# All packages will be installed here and survive session restarts.
# The original path '/content/drive/Shareddrives/AAI-590-Capstone-Project/colab_env' caused an OSError.
# We are changing it to a path within MyDrive, which is typically supported.
ENV_PATH = os.path.join(DRIVE_MOUNT, "MyDrive", "AAI-590-Capstone-Project", "colab_env")
os.makedirs(ENV_PATH, exist_ok=True)

# Add the environment path to sys.path so Python can find installed packages
if ENV_PATH not in sys.path:
    sys.path.insert(0, ENV_PATH)

print(f"Google Drive mounted at : {DRIVE_MOUNT}")
print(f"Persistent env path     : {ENV_PATH}")
print(f"sys.path[0]             : {sys.path[0]}")
print("Google Drive mounted and environment path configured.")
````

### 5) Install Dependencies from `requirements.txt`

Packages are installed from the `requirements.txt`.

````python
REQUIREMENTS_PATH = os.path.join(REPO_DIR, "requirements.txt")

if not os.path.isfile(REQUIREMENTS_PATH):
    raise FileNotFoundError(
        f"requirements.txt not found at {REQUIREMENTS_PATH}. "
        "Ensure it is committed to the root of the repository."
    )

print("Installing packages from requirements.txt ...\n")
print("─" * 60)
with open(REQUIREMENTS_PATH) as f:
    print(f.read())
print("─" * 60)

# --quiet suppresses per-package download noise; remove if you want verbose output
run(f"pip install --upgrade -r {REQUIREMENTS_PATH}")

print("\n All packages installed/verified.")
````

## 6) Configure the Python Path and Directory Structure

Adds `src/` to the Python path so all notebooks can import from `src/data_loader.py`, `src/preprocessing.py`, etc., without modification.

````python
import sys

# ── Path configuration ────────────────────────────────────────────────────────
SRC_PATH       = os.path.join(REPO_DIR, "src")
NOTEBOOKS_PATH = os.path.join(REPO_DIR, "notebooks")

for path in [REPO_DIR, SRC_PATH, NOTEBOOKS_PATH]:
    if path not in sys.path:
        sys.path.insert(0, path)

print("Python path updated:")
for p in sys.path[:5]:
    print(f"  {p}")

# ── Local data/output directories (gitignored, created on demand) ─────────────
LOCAL_DIRS = [
    os.path.join(REPO_DIR, "data", "raw"),
    os.path.join(REPO_DIR, "data", "processed"),
    os.path.join(REPO_DIR, "reports", "figures"),
]

for d in LOCAL_DIRS:
    os.makedirs(d, exist_ok=True)

print("\nLocal directories verified:")
for d in LOCAL_DIRS:
    print(f"  {d}")

print("\n Path and directory setup complete.")
````

## 7) Verify Core Imports

Confirms that all key libraries installed correctly and are importable.

````python
import importlib

# Map: display name → importable module name
PACKAGES = {
    "pandas":                 "pandas",
    "numpy":                  "numpy",
    "matplotlib":             "matplotlib",
    "seaborn":                "seaborn",
    "scikit-learn":           "sklearn",
    "statsmodels":            "statsmodels",
    "prophet":                "prophet",
    "torch (PyTorch)":        "torch",
    "pytorch-forecasting":    "pytorch_forecasting",
    "pytorch-lightning":      "lightning",
    "requests":               "requests",
}

all_ok = True
print(f"{'Package':<28} {'Version':<15} Status")
print("─" * 55)

for display_name, module_name in PACKAGES.items():
    try:
        mod = importlib.import_module(module_name)
        version = getattr(mod, "__version__", "n/a")
        print(f"{display_name:<28} {version:<15} ✅")
    except ImportError as e:
        print(f"{display_name:<28} {'':15} ❌  {e}")
        all_ok = False

print()
if all_ok:
    print("All packages imported successfully.")
else:
    print("One or more packages failed to import. Re-run Step 4 or restart the runtime.")
````
